```python
import pandas as pd

# Assuming your dataset is loaded into a DataFrame called df
# df = pd.read_csv('pizza_sales_data.csv')  # Example

# Total Revenue
total_revenue = df['total_price'].sum()

# Total Orders
total_orders = df['order_id'].nunique()

# Total Pizzas Sold
total_pizzas_sold = df['quantity'].sum()

# Average Order Value (AOV)
average_order_value = total_revenue / total_orders

# Average Pizza per Order
average_pizza_per_order = total_pizzas_sold / total_orders

# Displaying the KPIs
print(f"Total Revenue: {total_revenue}")
print(f"Total Orders: {total_orders}")
print(f"Total Pizzas Sold: {total_pizzas_sold}")
print(f"Average Order Value: {average_order_value}")
print(f"Average Pizza per Order: {average_pizza_per_order}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataset is loaded into a DataFrame called df
# df = pd.read_csv('pizza_sales_data.csv')

# Ensure 'order_date' is in datetime format and 'day_of_week' is created
df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y')
df['day_of_week'] = df['order_date'].dt.day_name()

# Group by 'day_of_week' and count the total number of orders
orders_by_day = df.groupby('day_of_week')['order_id'].nunique().sort_index()

# Define the order of days for the x-axis
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=orders_by_day.index, y=orders_by_day.values, order=day_order, palette='viridis')

# Customizing the chart
plt.title('Total Orders by Day of the Week', fontsize=16)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Total Orders', fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Assuming your dataset is loaded into a DataFrame called df
# df = pd.read_csv('pizza_sales_data.csv')

# Ensure 'order_date' is in datetime format
df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y')

# Extract month and year for grouping
df['year_month'] = df['order_date'].dt.to_period('M')

# Group by 'year_month' and count the total number of orders
monthly_orders = df.groupby('year_month')['order_id'].nunique()

# Convert 'year_month' to string for plotting
monthly_orders.index = monthly_orders.index.astype(str)

# Plotting the line chart
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_orders.index, y=monthly_orders.values, marker='o', color='blue')

# Customizing the chart
plt.title('Monthly Total Orders', fontsize=16)
plt.xlabel('Month-Year', fontsize=12)
plt.ylabel('Total Orders', fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Assuming your dataset is loaded into a DataFrame called df
# df = pd.read_csv('pizza_sales_data.csv')

# Ensure 'total_price' is numeric
df['total_price'] = pd.to_numeric(df['total_price'], errors='coerce')

# Group by 'pizza_category' and sum the total sales
sales_by_category = df.groupby('pizza_category')['total_price'].sum()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sales_by_category, labels=sales_by_category.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(sales_by_category))))

# Customizing the chart
plt.title('Distribution of Sales Across Different Pizza Categories', fontsize=16)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Assuming your dataset is loaded into a DataFrame called df
# df = pd.read_csv('pizza_sales_data.csv')

# Group by 'pizza_category' and sum the total quantity sold
quantity_by_category = df.groupby('pizza_category')['quantity'].sum().sort_values(ascending=False)

# Define the number of categories
n_categories = len(quantity_by_category)

# Define the position of bars to create the funnel effect
positions = np.arange(n_categories)
widths = np.linspace(1, 0.5, n_categories)  # Funnel effect, decreasing width

# Plotting the funnel chart
plt.figure(figsize=(10, 8))
bars = plt.barh(positions, quantity_by_category, height=widths, color='skyblue', edgecolor='black')

# Customizing the chart
plt.title('Funnel Chart: Number of Pizzas Sold for Each Pizza Category', fontsize=16)
plt.xlabel('Number of Pizzas Sold', fontsize=12)
plt.yticks(positions, quantity_by_category.index)
plt.gca().invert_yaxis()  # Invert y-axis to have the largest value on top
plt.grid(axis='x')

# Show the plot
plt.tight_layout()
plt.show()
```
